```python
import numpy as np
from scipy.stats import norm
```

```python
sample = [172, 174, 168, 169, 171, 173, 175, 170, 169, 172]
population_mean= 170
std = 3 
samp_mean = np.mean(sample)
n = len(sample)
```

```python
z_score = (samp_mean - population_mean)/(std/np.sqrt(n))
z_score
```

```python
p_value = 2 * (1 - norm.cdf(abs(z_score)))
p_value
```

```python
alpha = 0.05  # significance level is either given in question or we take it as 0.05
if p_value<alpha:
    print("Reject null hypothesis")
else: 
    print("accept Null Hypothesis")
```

### T Test


```python
import numpy as np
from scipy import stats
```

```python
sample = [172, 174, 168, 169, 171, 173, 175, 170, 169, 172]
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1) # In T-test ddof is generally 1.
n = len(sample)
mean_population = 170
```

```python
t_stats = (sample_mean-population_mean)/(sample_std/np.sqrt(n))
t_stats
```

```python
p_value = 2 * (1 - stats.t.cdf(abs(t_stats),df=n-1))
p_value
```

```python
alpha = 0.05
if p_value<alpha:
    print("Reject Null Hypothesis")
else:
    print("Accept Null Hypothesis")
```

### Two Sample T Test


```python

import numpy as np
from scipy import stats
```

```python
group_A = [85, 88, 90, 92, 87, 85, 89, 91, 86, 88]
group_B = [82, 84, 80, 83, 81, 79, 78, 85, 84, 83]
```

```python
t_value, p_value = stats.ttest_ind(group_A,group_B,equal_var=False)
```

```python
t_value
```

```python
alpha = 0.05
if p_value < alpha:
    print("Reject the Null Hypothesis")
else:
    print("Accept the Null Hypothesis")
```

### Chi Square Test

Works on Categorical Data

```python
import numpy as np
import seaborn as sns
from scipy.stats import chi2_contingency
import pandas as pd
```

```python
df = sns.load_dataset("titanic")
df.sample(2)
```

```python
contingency_table = pd.crosstab(df['sex'],df['survived'])
contingency_table
```

```python
chi2, p_value, dof, expected = chi2_contingency(contingency_table)
```

```python
expected
```

```python
alpha = 0.05
if p_value<alpha:
    print("Reject Null Hypothesis and there is a significance relation between Gender and survival")
else: 
    print("Accept Null Hypothesis and there is no significance relation between Gender and Survival")
    
```

### One Way ANOVA


```python
import numpy as np
import seaborn as sns
from scipy.stats import f_oneway
```

```python
df = sns.load_dataset("titanic")
```

```python
df.sample(2)
```

```python
df = df[['pclass','age']].dropna()
```

```python
df["pclass"].unique()
```

```python
class_1 = df[df['pclass']==1]['age']
class_2 = df[df['pclass']==2]['age']
class_3 = df[df['pclass']==3]['age']
```

```python

f_stats, p_value = f_oneway(class_1, class_2, class_3)
```

```python
f_stats
```

```python
alpha = 0.05
if p_value < alpha:
    print("Reject the Null Hypothesis and there is a sifnificance dfference between atleast one pclass")
else:
    print("Accept the Null Hypothesis and there is no difference")
```
