```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df = sns.load_dataset("iris")
df.sample(5)
```

```python
df.rename(columns={'sepal_length':'sl','sepal_width':'sw','petal_length':'pl','petal_width':'pw'},inplace=True)
```

```python
df.sample(5)
```

```python
setosa = df[df['species'] == 'setosa']
versicolor = df[df['species'] == 'versicolor']
virginica = df[df['species'] == 'virginica']

```

```python
sns.displot(x = setosa['pl'], kde = True)
```

```python
sns.FacetGrid(df, hue='species', palette='cool',height=6).map(sns.histplot,'pl', kde=True).add_legend()
```

```python
sns.FacetGrid(df, hue='species', palette='cool',height=6).map(sns.histplot,'pw', kde=True).add_legend()
```

```python
sns.FacetGrid(df, hue='species', palette='cool',height=6).map(sns.histplot,'sl', kde=True).add_legend()
```

```python
sns.FacetGrid(df, hue='species', palette='cool',height=6).map(sns.histplot,'sw', kde=True).add_legend()
```

```python
plt.figure(figsize=(6,6))
sns.scatterplot(virginica['pw'], color = 'Blue')
sns.scatterplot(versicolor['pw'], color = 'Green')
sns.scatterplot(setosa['pw'], color = 'Red')
plt.legend(['virginica','versicolor', 'setosa'])
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.scatterplot(virginica['pl'], color = 'Blue')
sns.scatterplot(versicolor['pl'], color = 'Green')
sns.scatterplot(setosa['pl'], color = 'Red')
plt.legend(['virginica','versicolor', 'setosa'])
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.scatterplot(virginica['sw'], color = 'Blue')
sns.scatterplot(versicolor['sw'], color = 'Green')
sns.scatterplot(setosa['sw'], color = 'Red')
plt.legend(['virginica','versicolor', 'setosa'])
plt.show()
```
