```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df = sns.load_dataset("iris")
```

```python
df
```

```python
df.head(10)
```

```python
df.rename(columns={'sepal_length':'sl','sepal_width':'sw','petal_length':'pl','petal_width':'pw'},inplace=True)
df.sample(5)
```

```python
df.species.value_counts()
```

```python
setosa = df[df['species']=='setosa']
versicolor = df[df['species']=='versicolor']
virginica = df[df['species']=='virginica']
```

```python
print(setosa.shape)
print(versicolor.shape)
print(virginica.shape)
```

```python
print(setosa['sl'].mean())
print(setosa['sw'].mean())
print(setosa['pl'].mean())
print(setosa['pw'].mean())
```

```python
print(setosa['sl'].median())
print(setosa['sw'].median())
print(setosa['pl'].median())
print(setosa['pw'].median())
```

```python
print(stats.mode(setosa['sl']))
print(stats.mode(setosa['sw']))
print(stats.mode(setosa['pl']))
print(stats.mode(setosa['pw']))
```
