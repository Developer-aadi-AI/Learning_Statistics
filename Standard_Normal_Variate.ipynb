```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

```python
# Generate 100000 values with mean around 1000 and Standard Deviation around 200
draw = norm.ppf(np.random.random(100000), loc = 1000, scale = 200).astype(int)
sns.displot(draw, kde = True )
```

```python
z  = []
for i in draw:
    z.append((i-1000)/200)

```

```python
a = np.array(z)
```

```python
print(a.mean())
print(np.round(a.mean()))
```

```python
print(a.std())
print(np.round(a.std()))
```

```python
sns.displot(a, kde = True)
```
